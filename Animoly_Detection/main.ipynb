```python
with open("data/kddcup.names") as f:
    names = f.readlines()
```

```python
col_names = []
for x in names:
    col_names.append(x.split(":")[0])
```

```python
col_names.append("target")
col_names.pop(0)
```

```python
len(names)
```

```python
df = pd.read_csv("data/kddcup.data_10_percent.gz", header=None, names=col_names)
```

```python
attacks_types = {
'normal': 'normal',
'back': 'dos',
'buffer_overflow': 'u2r',
'ftp_write': 'r2l',
'guess_passwd': 'r2l',
'imap': 'r2l',
'ipsweep': 'probe',
'land': 'dos',
'loadmodule': 'u2r',
'multihop': 'r2l',
'neptune': 'dos',
'nmap': 'probe',
'perl': 'u2r',
'phf': 'r2l',
'pod': 'dos',
'portsweep': 'probe',
'rootkit': 'u2r',
'satan': 'probe',
'smurf': 'dos',
'spy': 'r2l',
'teardrop': 'dos',
'warezclient': 'r2l',
'warezmaster': 'r2l',
}
```

```python
# df['target'].value_counts()
```

```python
#mapping the attack types based on target 
df['Attack Type'] = df.target.apply(lambda r:attacks_types[r[:-1]])
```

```python
df['Attack Type'].value_counts()
```

```python
df = df.drop('attack_type',axis=1)
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
len(cat_cols), len(num_cols) 
```

### Categorical Encoding

```python
from sklearn.preprocessing import StandardScaler    
```

```python
data = pd.get_dummies(df, columns=cat_cols).astype(int)
# data.head()

scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
```

## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_forest = IsolationForest(contamination=0.1, random_state=42)
```

```python
data['Anomaly'] = iso_forest.fit_predict(data)

anamoaly = data[data['Anomaly'] == -1]
```

```python
data['Anomaly'].value_counts()
#1 : normal
#-1 : anomaly
```

```python
# Select a subset of features for the pair plot
features = ['duration', 'src_bytes', 'dst_bytes', 'count', 'srv_count', 'Anomaly']

# Create a pair plot
sns.pairplot(data[features], hue='Anomaly', palette='coolwarm', markers=["o", "s"])
plt.suptitle('Pair Plot for Anomaly Detection', y=1.02)
plt.show()
```

```python
sorted_data = data.sort_values(by='duration')

# Plot 'duration' vs 'src_bytes' as a line plot
plt.figure(figsize=(12, 6))
plt.plot(sorted_data['duration'], sorted_data['src_bytes'], label='Normal Data')

# Highlight anomalies
anomalies = sorted_data[sorted_data['Anomaly'] == 1]
plt.scatter(anomalies['duration'], anomalies['src_bytes'], color='red', label='Anomalies')

plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel('Duration')
plt.ylabel('Source Bytes')
plt.legend()
plt.show()
```

## AutoEncoder


```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
```

```python
X = data.values

X_scaled = scaler.fit_transform(X)

X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
```

```python
input_dim = X_train.shape[1]
encdoing_dm = 14

input_layer = Input(shape=(input_dim,))
encoder = Dense(encdoing_dm, activation='relu')(input_layer)
decoder = Dense(input_dim, activation='sigmoid')(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mse')

history = autoencoder.fit(X_train, X_train, epochs=10, batch_size=32, validation_split=0.2, shuffle=True)

X_test_pred = autoencoder.predict(X_test)

```

```python
X_test_pred = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - X_test_pred, 2), axis=1)
threshold = np.percentile(mse, 95)  # Set threshold as 95th percentile of MSE

# Classify anomalies
anomalies = mse > threshold
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(len(mse)), mse, label='Reconstruction Error')
plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
plt.title('Anomaly Detection using Autoencoder')
plt.xlabel('Sample Index')
plt.ylabel('Reconstruction Error')
plt.legend()
plt.show()
```
